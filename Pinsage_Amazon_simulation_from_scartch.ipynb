```python
from huggingface_hub import hf_hub_download
import json
from collections import defaultdict
from tqdm import tqdm
```

```python

review_path = hf_hub_download(
    repo_id="McAuley-Lab/Amazon-Reviews-2023",
    filename="raw/review_categories/Electronics.jsonl",
    repo_type="dataset"
)
```

```python
meta_path = hf_hub_download(
    repo_id="McAuley-Lab/Amazon-Reviews-2023",
    filename='raw/meta_categories/meta_Electronics.jsonl',
    repo_type="dataset"
)
```

```python
with open(review_path, 'r') as f:
    samples = [json.loads(f.readline()) for _ in range(10)]


type(samples[0])
for sample in samples:
  for i in list(sample.keys()):
    print(f'{i}: {sample[i]}')
  print()
    # total = sum(1 for _ in f)
```

```python
with open(meta_path, 'r') as f:
    metas = [json.loads(f.readline()) for _ in range(10)]

for sample in metas:
  for i in list(sample.keys()):
    print(f'{i}: {sample[i]}')
  print()
```

```python
user_buy = defaultdict(set)
with open(review_path,'r') as f:
  for line in tqdm(f,desc='one'):
    buy = json.loads(line)
    if buy.get('verified_purchase'):
      user_buy[buy['user_id']].add(buy['parent_asin'])


```

```python
print(len(user_buy))
```

```python
user_buy_r = defaultdict(set)
with open(review_path,'r') as f:
  for line in tqdm(f,desc='one'):
    buy = json.loads(line)
    user_buy_r[buy['user_id']].add(buy['parent_asin'])
```

```python
print(len(user_buy_r))
```
